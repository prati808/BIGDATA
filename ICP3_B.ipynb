Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
```

```python
car_sales['Price'] = car_sales['Price'].astype(str)

# Now remove $ and commas using raw string r'' for regex pattern
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$,]', '', regex=True)

# Remove the last two characters
car_sales['Price'] = car_sales['Price'].str[:-2]

# Convert to numeric, coercing errors to NaN just in case
car_sales['Price'] = pd.to_numeric(car_sales['Price'], errors='coerce')
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame
```

```python
from datetime import datetime

# Add 'Total Sales' as the cumulative sum of 'Price'
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add 'Sale Date' starting from today, one date per row
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp(datetime.today().date()), periods=len(car_sales), freq='D')

# View the updated DataFrame
print(car_sales.head())
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot


# Plot the data


# Customize the plot


# Add a meanline
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
axes = axes.flatten()

# Select numeric columns (limit to 9 for this example)
numeric_cols = over_50.select_dtypes(include='number').columns[:9]

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    
    # Scatter plot with 'winter' colormap
    scatter = ax.scatter(over_50.index, over_50[col], c=over_50[col], cmap='winter')
    
    # Add red horizontal mean line
    mean_val = over_50[col].mean()
    ax.axhline(mean_val, color='red', linestyle='dashed', linewidth=1)
    
    ax.set_title(f'{col} (mean: {mean_val:.2f})')
    ax.set_xlabel('Index')
    ax.set_ylabel(col)
    
    # Add colorbar for scatter plot
    fig.colorbar(scatter, ax=ax)

# Remove unused subplots if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
